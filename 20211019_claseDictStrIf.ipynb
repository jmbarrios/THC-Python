## Diccionarios (_HashMap_)

```python
col_names = ['datetime', 'scientific_name', 'species', 'island', 'continent', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex', 'recorded_by']
```

```python
print(col_names)
```

```python
print(list(zip(col_names, record)))
```

```python
record_dict = dict(zip(col_names, record))
```

```python
print(record_dict)
```

```python
my_exciting_feature = record_dict['bill_length_mm']*record_dict['bill_depth_mm']
print(my_exciting_feature)
```

```python
record_dict['age_years'] = 7
print(record_dict)
```

```python
my_exciting_feature = record_dict['bill_length_mm']*record_dict['bill_depth_mm']
print(my_exciting_feature)
```

```python
inspect(record_dict, methods=True)
```

```python
record2_dict = dict(
    zip(
        col_names, 
        records_w_date[1].copy()
        )
    )
print(record2_dict)
```
